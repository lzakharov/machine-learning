```python
import numpy as np
import pandas as pd
from sklearn import linear_model
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
%matplotlib inline
```

## Loading data

```python
city = 'Chengdu'
file = city + 'PM20100101_20151231.csv'

df = pd.read_csv('./data/' + file, usecols=['TEMP', 'DEWP', 'HUMI']).dropna()

df.head()
```

```python
plt.plot(df['TEMP'], df['DEWP'], '.')
```

## Selecting data

```python
X = df['TEMP'].values.reshape(-1, 1)
y = df['DEWP'].values.reshape(-1, 1)
```

## Training Ridge Regression

```python
alphas = [0.01, 0.1, 0.5]

for alpha in alphas:
    ridge = linear_model.Ridge(alpha=alpha, normalize=True)
    ridge.fit(X, y)
    y_pred = ridge.predict(X)
    print(alpha, ridge.coef_)
    plt.plot(X, y_pred)
    plt.plot(X, y, '.')
    plt.show()
```

## Training LASSO Regression

```python
alphas = [0.01, 0.1, 0.5]

for alpha in alphas:
    lasso = linear_model.Lasso(alpha=alpha, normalize=True)
    lasso.fit(X, y)
    y_pred = lasso.predict(X)
    print(alpha, lasso.coef_)
    plt.plot(X, y_pred)
    plt.plot(X, y, '.')
    plt.show()
```

## Training ElasticNet Regression

```python
alphas = [0.01, 0.1, 0.5]

for alpha in alphas:
    elastic_net = linear_model.ElasticNet(alpha=alpha, l1_ratio=0.7)
    elastic_net.fit(X, y)
    y_pred = elastic_net.predict(X)
    plt.title(elastic_net.coef_)
    plt.plot(X, y_pred)
    plt.plot(X, y, '.')
    plt.show()
```
