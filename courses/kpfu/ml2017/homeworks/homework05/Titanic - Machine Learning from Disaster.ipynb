```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train = pd.read_csv('train.csv')
train.fillna(train.median(), inplace=True)
train.dropna(inplace=True)
test = pd.read_csv('test.csv')
test.fillna(train.median(), inplace=True)

all_data = train.append(test)
all_data.head()
```

```python
all_data['Sex'] = all_data['Sex'].map({'male': 0, 'female': 1})
all_data['Embarked'] = all_data['Embarked'].map({'C': 0, 'Q': 1, 'S': 2})
all_data.drop('PassengerId', axis=1, inplace=True)
all_data.drop('Name', axis=1, inplace=True)
all_data.drop('Cabin', axis=1, inplace=True)
all_data.drop('Ticket', axis=1, inplace=True)

all_data.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import Normalizer
from sklearn.neural_network import MLPClassifier

X_train = all_data[:len(train)].drop('Survived', axis=1)
y_train = all_data[:len(train)]['Survived']

mlp = MLPClassifier(hidden_layer_sizes=(100, 100), max_iter=1000, alpha=0.01,
                    solver='sgd', tol=1e-4)

mlp.fit(X_train, y_train)
mlp.score(X_train, y_train)
```

```python
X_test = all_data[len(train):].drop('Survived', axis=1)
prediction = mlp.predict(X_test)

df_prediction = pd.DataFrame({'PassengerId': test['PassengerId'], 'Survived': prediction.astype(int)}).set_index('PassengerId')
df_prediction.to_csv('prediction.csv')
```
